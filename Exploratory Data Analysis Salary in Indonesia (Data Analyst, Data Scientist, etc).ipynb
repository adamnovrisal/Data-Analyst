```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd = pd.read_csv('Gaji.csv', encoding='cp1252')
```

```python
pd
```

```python
pd = pd.drop('Unnamed: 0', axis=1)
```

```python
pd
```

```python
#Little cleaning, removing "Lab/Chemical Analyst" and "Advertiser" because it's not related with Data Analyst

pd = pd[~pd['Title'].str.contains(r'lab|laboratorium|advertiser|chemical',regex=True,case=False)]
```

```python
pd
```

```python
#let's group the job title into specific type of job related to data (Data Analyst, Data Scientist, etc)

job_data = []

for job in pd.Title:
    if 'analyst' in job.lower():
        job_data.append('Data Analyst')
    elif 'scien' in job.lower():
        job_data.append('Data Scientist')
    elif 'develop' in job.lower():
        job_data.append('Developer')
    elif 'engineer' in job.lower():
        job_data.append('Data Engineer')
    elif 'IT' in job.lower():
        job_data.append('IT Specialist')
    elif 'program' in job.lower():
        job_data.append('Programmer')
    else:
        job_data.append('Others')
```

```python
#To show unique categories and the occurence number

np.unique(job_data, return_counts=True)
```

```python
pd['Data Job Type']=job_data
```

```python
pd
```

```python
#plot a bar to show salary based on job type

title_dict = {'fontsize': 30,
 'fontweight' : 'bold',}
label_dict = {'fontsize': 15,
 'fontweight' : 'normal',
 }
stitle_dict = {'fontsize': 20,
 'fontweight' : 'bold',}

plt.figure(figsize=(15,9))
ax = sns.countplot(x = 'Data Job Type', data=pd)
ax.set_title('Employee by Data Job Type', fontdict = label_dict);
ax.set_xlabel('Job Type', fontdict = label_dict);
ax.set_ylabel('Total Employee', fontdict = label_dict);

plt.figure(figsize=(15,9))
ax = sns.countplot(x = 'Experience Years Needed', data=pd)
ax.set_xlabel('Experience Level', fontdict = label_dict);
ax.set_ylabel('Total Employee', fontdict = label_dict);

plt.figure(figsize=(15,9))
ax = sns.countplot(x = 'Job Level', data=pd)
ax.set_xlabel('Job Level', fontdict = label_dict);
ax.set_ylabel('Total Employee', fontdict = label_dict);
```

```python
# First, we need to get rid table that don't have salary (-1), and salary with SGD currency

pd_salary = pd[~pd['Salary'].str.contains(r'-1|SGD')].reset_index()

# Next, we'll only extract the numeric content and place it in new column

# Get rid of "IDR", " per bulan", " jt" and whitespace (" ") by replacing them with empty string ("")
pd_salary['Salary'] = pd_salary['Salary'].apply(lambda x: x.replace('IDR ','').replace(' per bulan','').replace(' jt','').replace(' ',''))

# Get the first number as Min Salary Range, and second as Max Salary Range (They're divided by "-" symbol by the way, so split them will do the trick)
pd_salary['Min Salary Range'] = pd_salary['Salary'].apply(lambda x: x.split('-')[0].replace(',','.'))
pd_salary['Max Salary Range'] = pd_salary['Salary'].apply(lambda x: x.split('-')[1].replace(',','.'))

# Change dtype from string into float (we use float instead of int because there are some Salary with decimal)
pd_salary[['Min Salary Range','Max Salary Range']] = pd_salary[['Min Salary Range','Max Salary Range']].astype(float)
```

```python
# Let's see the basic stats of our Min Salary and Max Salary
pd_salary[['Min Salary Range','Max Salary Range']].describe()
```

```python
pd_salary
```

```python
# Removing "pekerja" that contained in the ['Company Scale]
pd_salary['Company Scale'] = pd_salary['Company Scale'].apply(lambda x: x.replace(r'pekerja','').replace(r'Lebih dari ','> '))
pd_salary = pd_salary[~pd_salary['Company Scale'].str.contains('None')]

# Show the average of max salary by each company scale
pd_max_salary_grouped_compscale = pd_salary.groupby('Company Scale',as_index=False)[['Max Salary Range']].mean().sort_values(by=['Max Salary Range'],ascending=True)
pd_max_salary_grouped_compscale['Max Salary Range'] = pd_max_salary_grouped_compscale['Max Salary Range'].round(2)
pd_max_salary_grouped_compscale
```

```python
# Plot the result

comp_scale = pd_max_salary_grouped_compscale['Company Scale']
salary = pd_max_salary_grouped_compscale['Max Salary Range']

plt.figure(figsize=(8, 6), dpi=80)
plt.barh(comp_scale,salary,height=0.65, color='maroon')
plt.xlabel('Average Salary (Million IDR)')
plt.xlim(0,max(salary)+3)
plt.ylabel('Company Scale (Number of Employees)')
plt.title('Average Salary based on Company Scale')
plt.grid()
# put number in top of bar
for i, v in enumerate(salary): # i is the iteration number, v is the content of y itself
    # plt.text(xticks = i, yticks = v+5(+5 will give a space between number and bar, the rest is font styling))
    plt.text(v+.5 , i, str(v), color='black', fontweight='normal', fontdict={'size' : 12, 'family': 'sans-serif'}, horizontalalignment = 'center')
plt.show()
```

- The above result seems quite shocking, showing that the bigger the company is not always guarantee a bigger salary you would get

- This raises new question.. Why is the result being like that?

- Hypothesis : Small company tend to hire small number of employees,but the person has to be an all-rounder (have various skills). That's why they dare to pay more.

- We'll test the hypothesis by combining the average salary with number of skills needed (please note that the skills are the one we already did above)

```python
# Removing "tahun" that contained in the ['Company Scale]
pd_salary['Experience Years Needed'] = pd_salary['Experience Years Needed'].apply(lambda x: x.replace(r'tahun','').replace(r'Lebih dari ','> '))
pd_salary = pd_salary[~pd_salary['Experience Years Needed'].str.contains('None')]

# Show the average of max salary by each company scale
pd_max_salary_grouped_experience = pd_salary.groupby('Experience Years Needed',as_index=False)[['Max Salary Range']].mean().sort_values(by=['Max Salary Range'],ascending=True)
pd_max_salary_grouped_experience['Max Salary Range'] = pd_max_salary_grouped_experience['Max Salary Range'].round(2)
pd_max_salary_grouped_experience
```

```python
# Plot the result

comp_scale = pd_max_salary_grouped_experience['Experience Years Needed']
salary = pd_max_salary_grouped_experience['Max Salary Range']

plt.figure(figsize=(8, 6), dpi=80)
plt.barh(comp_scale,salary,height=0.65, color='maroon')
plt.xlabel('Average Salary (Million IDR)')
plt.xlim(0,max(salary)+3)
plt.ylabel('Experience Years Needed (Number of Employees)')
plt.title('Average Salary based on Experience Years Needed')
plt.grid()
# put number in top of bar
for i, v in enumerate(salary): # i is the iteration number, v is the content of y itself
    # plt.text(xticks = i, yticks = v+5(+5 will give a space between number and bar, the rest is font styling))
    plt.text(v+.5 , i, str(v), color='black', fontweight='normal', fontdict={'size' : 12, 'family': 'sans-serif'}, horizontalalignment = 'center')
plt.show()
```

```python
# Removing "hari" that contained in the ['Company Scale]
pd_salary['Processed Time'] = pd_salary['Processed Time'].apply(lambda x: x.replace(r'hari','').replace(r'None ','> '))
pd_salary = pd_salary[~pd_salary['Processed Time'].str.contains('None')]

# Show the average of max salary by each company scale
pd_max_salary_grouped_process = pd_salary.groupby('Processed Time',as_index=False)[['Max Salary Range']].mean().sort_values(by=['Max Salary Range'],ascending=True)
pd_max_salary_grouped_process['Max Salary Range'] = pd_max_salary_grouped_process['Max Salary Range'].round(2)
pd_max_salary_grouped_process
```

```python
# Plot the result

process = pd_max_salary_grouped_process['Processed Time']
salary = pd_max_salary_grouped_process['Max Salary Range']

plt.figure(figsize=(8, 6), dpi=80)
plt.barh(process,salary,height=0.65, color='maroon')
plt.xlabel('Average Salary (Million IDR)')
plt.xlim(0,max(salary)+3)
plt.ylabel('Processed Time (Number of Employees)')
plt.title('Average Salary based on Processed Time')
plt.grid()
# put number in top of bar
for i, v in enumerate(salary): # i is the iteration number, v is the content of y itself
    # plt.text(xticks = i, yticks = v+5(+5 will give a space between number and bar, the rest is font styling))
    plt.text(v+.5 , i, str(v), color='black', fontweight='normal', fontdict={'size' : 12, 'family': 'sans-serif'}, horizontalalignment = 'center')
plt.show()
```
